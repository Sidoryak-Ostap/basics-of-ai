#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

### **Нелінійних рівнянь Нарендра-Партхізаратхі:**

\dot{x}_1 = -a_1 x_1 + b_1 x_2^2
$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \sin(x_1)
$$

2. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2
$$
$$
\dot{x}_2 = a_2 x_1 x_2 - b_2 \cos(x_1)
$$

3. $$

$$
$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2

$$
$$
\dot{x}_2 = -a_2 x_2 + b_2 \cos(x_1)
$$

5. $$
\dot{x}_1 = -a_1 x_1 + b_1 x_2^2 \cos(x_1)
$$
$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_1 x_2

6. $$

$$
\dot{x}_2 = a_2 \sin(x_1) - b_2 x_2^2

7. $$
\dot{x}_1 = -a_1 x_1^2 + b_1 \cos(x_2)
$$
$$
\dot{x}_2 = a_2 x_1 - b_2 \sin(x_2)

8. $$
\dot{x}_1 = a_1 \sin(x_2) - b_1 x_1^2

$$
\dot{x}_2 = -a_2 x_2 + b_2 x_1 \cos(x_2)

9. $$
\dot{x}_1 = a_1 x_1^2 - b_1 x_2^2

$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 \sin(x_2)

10. $$
\dot{x}_1 = -a_1 x_1 \sin(x_2) + b_1 x_2^3
$$
$$
\dot{x}_2 = a_2 \cos(x_1) - b_2 x_1 x_2

11. $$
\dot{x}_1 = a_1 \sin(x_1 x_2) - b_1 x_1

$$

### Параметри:
- $( a_1, a_2 )$, $( b_1, b_2 )$ — це коефіцієнти, які можуть бути варійовані для дослідження динаміки системи та впливу на прогнозування.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import svm
from sklearn.metrics import mean_squared_error, r2_score

```

```python

sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

args = (sigma, rho, beta)

def lorenz(X, t, sigma, rho, beta):
    x, y, z = X
    dxdt = sigma * (y - x)
    dydt = x * (rho - z) - y
    dzdt = x * y - beta * z
    return [dxdt, dydt, dzdt]

initial_conditions = [1, 1, 1]

t = np.linspace(0, 50, 10000)  

solution = odeint(lorenz, initial_conditions, t, args=args)

lorenz_objects_df = pd.DataFrame(solution, columns=['x','y','z'])

lorenz_objects_df.to_csv('./lorenz-data.csv')
lorenz_objects_df.head()
```

```python


a1 = 1.0
a2 = 2.0
b1 = 0.5
b2 = 0.8

def narendra_parthasarathy(X, t, a1, a2, b1, b2):
    x1, x2 = X
    dx1dt = -a1 * np.sin(x1) + b1 * (x2 ** 2)
    dx2dt = a2 * np.sin(x1 * x2) - b2 * np.cos(x1)
    return [dx1dt, dx2dt]

initial_conditions = [1.0, 1.0]

t = np.linspace(0, 50, 10000)  


solution = odeint(narendra_parthasarathy, initial_conditions, t, args=(a1, a2, b1, b2))

narendra_objects_df = pd.DataFrame(solution, columns=['x','y'])
narendra_objects_df.to_csv('./narendra.csv')
narendra_objects_df.head(5)
```

```python
train_lorenz_data, temp_lorenz_data = train_test_split(lorenz_objects_df, test_size=0.30, random_state=42)
val_lorenz_data, test_lorenz_data = train_test_split(temp_lorenz_data, test_size=0.50, random_state=42)

print("Size of train_lorenz_data:", len(train_lorenz_data))
print("Size of val_lorenz_data:", len(val_lorenz_data))
print("Size of test_lorenz_data:", len(test_lorenz_data))
```

```python

train_narendra_data, temp_narendra_data = train_test_split(narendra_objects_df, test_size=0.30, random_state=42)
val_narendra_data, test_narendra_data = train_test_split(temp_narendra_data, test_size=0.50, random_state=42)

print("Size of train_narendra_data:", len(train_narendra_data))
print("Size of val_narendra_data:", len(val_narendra_data))
print("Size of test_narendra_data:", len(test_narendra_data))


```

```python
scaler = StandardScaler()
```

Система Лоренца

```python
X_train = train_lorenz_data[['x', 'y']]
y_train = train_lorenz_data['z']

X_val = val_lorenz_data[['x', 'y']]
y_val = val_lorenz_data['z']

X_test = test_lorenz_data[['x', 'y']]
y_test = test_lorenz_data['z']

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100),
    "SVM": svm.SVR(C=30, kernel='rbf', gamma="auto")
}

results = {}

for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_val_pred = model.predict(X_val_scaled)
    
    mse = mean_squared_error(y_val, y_val_pred)
    r2 = r2_score(y_val, y_val_pred)
    
    results[name] = {"MSE": mse, "R2 Score": r2}

for model_name, metrics in results.items():
    print(f"{model_name} - MSE: {metrics['MSE']}, R2 Score: {metrics['R2 Score']}")
```

```python

model_params = {
    "Ridge Regression": {
        "model": Ridge(),
        "params": {'alpha': [0.1, 1.0, 10.0, 100.0]}
    },
    "Decision Tree": {
        "model": DecisionTreeRegressor(),
        "params": {
            'max_depth': [None, 5, 10, 15],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    "Random Forest": {
        "model": RandomForestRegressor(),
        "params": {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 5, 10, 15],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    "SVM": {
        "model": svm.SVR(),
        "params": {
            'C': [0.1, 1.0, 10.0, 100.0],
            'kernel': ['linear', 'poly', 'rbf'],
            'gamma': ['scale', 'auto']
        }
    }
}

for model_name, mp in model_params.items():
    grid = GridSearchCV(mp['model'], mp['params'], cv=5)
    grid.fit(X_train_scaled, y_train)
    
    print(f"Best parameters for {model_name}: {grid.best_params_}")
    print(f"Best score for {model_name}: {grid.best_score_}")
```

```python

best_models = {
    "Ridge Regression": model_params["Ridge Regression"]["model"],
    "Decision Tree":  model_params["Decision Tree"]["model"],
    "Random Forest":  model_params["Random Forest"]["model"],
    "SVM": model_params["SVM"]["model"]
}

results = {}

for model_name, model in best_models.items():
    model.fit(X_train_scaled, y_train)
    y_test_pred = model.predict(X_test_scaled)
    
    mse = mean_squared_error(y_test, y_test_pred)
    r2 = r2_score(y_test, y_test_pred)
    
    results[model_name] = {"MSE": mse, "R²": r2}

for model_name, metrics in results.items():
    print(f"Validation MSE for {model_name}:", metrics["MSE"])
    print(f"Validation R² for {model_name}:", metrics["R²"])
```

```python

def plot_predictions(model_name, y_test, y_test_pred):
    plt.figure(figsize=(8, 6))
    plt.scatter(y_test, y_test_pred, alpha=0.5)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--') 
    plt.xlabel("Actual Values")
    plt.ylabel("Predicted Values")
    plt.title(f"{model_name} - Actual vs Predicted Values")
    plt.show()

for model_name, best_model in best_models.items():
    best_model.fit(X_train_scaled, y_train)
    y_test_pred = best_model.predict(X_test_scaled)
    plot_predictions(model_name, y_test, y_test_pred)
```

```python

X_train = train_narendra_data[['x', 'y']]
y_train = train_narendra_data['y']  

X_val = val_narendra_data[['x', 'y']]
y_val = val_narendra_data['y']

X_test = test_narendra_data[['x', 'y']]
y_test = test_narendra_data['y']

X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)
```

```python

def train_and_evaluate(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test, predictions)
    r2 = r2_score(y_test, predictions)
    return mse, r2

models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100),
    "SVM": svm.SVR(C=30, kernel='rbf', gamma="auto")
}

results = {}
for model_name, model in models.items():
    mse, r2 = train_and_evaluate(model, X_train_scaled, y_train, X_test_scaled, y_test)
    results[model_name] = {"MSE": mse, "R2 Score": r2}

for model_name, metrics in results.items():
    print(f"{model_name}: MSE = {metrics['MSE']:.4f}, R2 Score = {metrics['R2 Score']:.4f}")
```

```python
def hyperparameter_tuning(model, params, X_train, y_train):
    grid_search = GridSearchCV(model, params, cv=5)
    grid_search.fit(X_train, y_train)
    return grid_search.best_params_, grid_search.best_score_

model_params = {
    "Ridge Regression": {
        "model": Ridge(),
        "params": {'alpha': [0.1, 1.0, 10.0, 100.0]}
    },
    "Decision Tree": {
        "model": DecisionTreeRegressor(),
        "params": {
            'max_depth': [None, 5, 10, 15],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    "Random Forest": {
        "model": RandomForestRegressor(),
        "params": {
            'n_estimators': [10, 50, 100],
            'max_depth': [None, 5, 10, 15],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    "SVM": {
        "model": svm.SVR(),
        "params": {
            'C': [0.1, 1.0, 10.0, 100.0],
            'kernel': ['linear', 'poly', 'rbf'],
            'gamma': ['scale', 'auto']
        }
    }
}

tuning_results = {}
for model_name, config in model_params.items():
    best_params, best_score = hyperparameter_tuning(config["model"], config["params"], X_train_scaled, y_train)
    tuning_results[model_name] = {"Best Params": best_params, "Best Score": best_score}

for model_name, result in tuning_results.items():
    print(f"{model_name}: Best Params = {result['Best Params']}, Best Score = {result['Best Score']:.4f}")
```
